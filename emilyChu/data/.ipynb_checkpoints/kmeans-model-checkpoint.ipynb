## a. Perform image processing
### (using hierarchical segmentation to define dominant colors and shapes via averaging)

```python
# in downscaling the image, what do you want the new dimensions to be?
dims = (140, 90)
```

```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pickle
import warnings
from skimage import io, color, transform, feature
from skimage.io import imread

from skimage.filters.rank import entropy
from skimage.morphology import disk
from skimage import filters
from scipy import ndimage as ndi
from skimage import morphology
from skimage import data, segmentation, filters, color
from skimage.future import graph
from skimage import exposure
from skimage.transform import rescale

from skimage.data import camera
from skimage.filters import roberts, sobel, scharr, prewitt
```

```python
def image_manipulation(imname, imview=False):
    warnings.filterwarnings('ignore')
    imname = images + imname + '.png'
    print(imname)
    img_raw = io.imread(imname)
    downscaled = transform.resize(img_raw, (dims[0], dims[1])) # downscale image
    gray = filters.sobel(color.rgb2gray(downscaled))
    edge_roberts = roberts(gray)
    edge_scharr = scharr(gray)
    edge_prewitt = prewitt(gray)
    edge_sobel = sobel(gray)
    diff_scharr_sobel = edge_roberts - edge_sobel

    if imview==True:
        io.imshow(diff_scharr_sobel)
    warnings.filterwarnings('always')
    return diff_scharr_sobel

# test the function, look at input/output
test_image = image_manipulation('1', True)
print('downscaled image shape: ')
print(test_image.shape)
print('image representation (first row of pixels): ')
print(test_image[0])
print('\n')
print('example of transformation: ')
original_test = io.imread(images+'1.png')
```

## b. Refine until transformed version of originals are satisfactory

```python
# Show original next to transformed
fig, ax = plt.subplots(ncols=2, sharex=False, sharey=False,
                       figsize=(10, 5))

ax[0].imshow(original_test, cmap=plt.cm.gray)
ax[0].set_title('Original')

ax[1].imshow(test_image, cmap=plt.cm.gray)
ax[1].set_title('Transformed')

plt.tight_layout()
plt.show()
```

## c. Save the image representations 

```python
# construct lists for features, labels, and a crosswalk reference to image names

features_list = []
y_list = []
imnames_list = []
objectid_list = []

for index, row in df.iterrows():
    try:
        features_list.append(image_manipulation(row['img']))
        y_list.append(row['timestamp'])
        imnames_list.append(row['img'])
    except OSError:
       continue
    
```

```python
# convert the lists to ndarrays
features = np.asarray(features_list)
Y = np.asarray(y_list)
imgs = np.asarray(imnames_list)
print('Shape of original feature representation: ')
print(features.shape)

# flatten the images ndarray to one row per image
features_flat = features.reshape((features.shape[0], -1))

print('Shape of flat feature representation: ')
print(features_flat.shape)

print('Shape of Y: ')
print(Y.shape)

print(features_flat)
print(imgs)
```

```python
# write to a csv
import csv
rows = zip(imgs, features)
with open("features.csv", "w") as f:
    writer = csv.writer(f)
    writer.writerow(['img','feature'])
    for row in rows:
        writer.writerow(row)
```

# run KMeans

```python
from sklearn import cluster

k_means = cluster.KMeans(n_clusters=5, n_jobs=-1)
k_means.fit(features_flat)
y_kmeans = k_means.predict(features_flat)
```

```python
labels = k_means.labels_
centroids = k_means.cluster_centers_
```

```python
import csv
rows = zip(centroids)
with open("centroids.csv", "w") as f:
    writer = csv.writer(f)
    writer.writerow(['centroids'])
    for row in rows:
        writer.writerow(row)
```
